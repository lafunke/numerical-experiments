```python
# create the FOM using ngsolve
# here define mesh, FE-space, operators

# here we tell ngsolve what to do when pyMOR calls it

import numpy as np
from ngsolve import *
from ngsolve.webgui import Draw
from ngsolve.meshes import Make1DMesh


alpha = 1.0
beta = Parameter(3.25)  # solve for different beta later
nu = 0.01
h = 50                  # (compare 50 and 100 later)
order = 2               # quadratic Lagrange basis functions

mesh = Make1DMesh(h)   # mesh on (0,1) into h pieces

# V_h = V_u * V_v 
V_h = H1(mesh, order=order, dirichlet="right") * H1(mesh, order=order, dirichlet="right")

u, v = V_h.TrialFunction()    # represent unknowns 
phi, psi = V_h.TestFunction() # test functions ("for u and v")


# This holds the values where the nonlinear operator is evaluated
w_h = GridFunction(V_h)          # w_h element in V_h with zeros
u_part, v_part = w_h.components  # seperate u and v part in w_h 

# generate mass Matrix m 
m = BilinearForm(V_h)           # m: V_h x V_h to R
m += (u * phi + v * psi) * dx   # u * phi fills the upper-left block (u part)
                                # v * psi fills the lower-right block (v part)
m.Assemble()

# generate stiffness matrix a
a = BilinearForm(V_h)          # a: V_h x V_h to R
a += (nu * grad(u) * grad(phi) + nu * grad(v) * grad(psi)) * dx
a.Assemble()

# H01-inner product
H01 = BilinearForm(V_h)
H01 += (grad(u) * grad(phi) + grad(v) * grad(psi)) * dx
H01.Assemble()

# for the non-lin part we use Newtons method F(u + \delta u) \approx F(u) + F'(u)* \delta u
# for Newton we need the residual and the Jacobian

# residual
res = LinearForm(V_h)

y_sym = alpha + u_part
z_sym = (beta / alpha) + v_part

res += (nu * grad(u_part) * grad(phi) + nu * grad(v_part) * grad(psi)) * dx   # diffusion part for both (grad(u_part)=grad(y_sym)
res += -(y_sym**2 * z_sym - (beta + 1) * y_sym + alpha) * phi * dx            # nonlinear part for y
res += -(beta * y_sym - y_sym**2 * z_sym) * psi * dx                          # nonlinear part for z

# jacobian
jac = BilinearForm(V_h)

jac += (nu * grad(u) * grad(phi) + nu * grad(v) * grad(psi)) * dx
jac += -( (2*y_sym*z_sym - (beta+1))*u + (y_sym**2)*v ) * phi * dx
jac += -( (beta - 2*y_sym*z_sym)*u - (y_sym**2)*v ) * psi * dx

# define the functions

def spatial_residual(vec, beta_val):
    
    beta.Set(beta_val.item() if hasattr(beta_val, 'item') else float(beta_val))
    w_h.vec.data = vec
    res.Assemble()
    
    return res.vec

def jacobian(vec, beta_val):
    beta.Set(beta_val.item() if hasattr(beta_val, 'item') else float(beta_val))
    w_h.vec.data = vec
   
    jac.Assemble()
    
    return jac.mat
```

```python
# bind the linear parts

from pymor.bindings.ngsolve import NGSolveMatrixOperator, NGSolveVectorSpace

# define the pyMOR VectorSpace from your NGSolve FE space
v_space = NGSolveVectorSpace(V_h)    # make it a pyMOR VectorSpace

m_op = NGSolveMatrixOperator(m.mat, range=v_space, source=v_space)   # pyMOR operator that takes VectorArray as input
a_op = NGSolveMatrixOperator(a.mat, range=v_space, source=v_space)

H01_op = NGSolveMatrixOperator(H01.mat, range=v_space, source=v_space)
```

```python
# here create the Operator A (we can plug into InstationaryModel)
# create a new Operator that inherits from Operator (pyMOR object)

from pymor.operators.interface import Operator

class BrusselatorOperator(Operator):
    def __init__(self, V_h, a_form, residual_func, jacobian_func):

        super().__init__()  # inherit from Operator

        self.linear = False
        
        self.V_h = V_h
        self.source = v_space
        self.range = v_space
        
        self.a_form = a_form
        self.residual_func = residual_func
        self.jacobian_func = jacobian_func

        self.parameters_own = {'beta': 1} # this is the parameter

    def apply(self, U, mu=None):

        gfu_impl = U.vectors[0].real_part.impl   # get the Gridfunction
    
        vec_ng = gfu_impl.vec    # get the BaseVector
        current_beta = mu['beta']   # update beta
    
        res_vec = self.residual_func(vec_ng, current_beta)  # compute residual in ngsolve, return basevector
    
        res_gf = GridFunction(self.V_h)  # wrap into new gridfunction
        res_gf.vec.data = res_vec
    
        return self.range.make_array([res_gf])      # make VectorArray

        
    def jacobian(self, U, mu=None):
        
        gfu_impl = U.vectors[0].real_part.impl
        vec_ng = gfu_impl.vec
        
        # compute jacobian in ngsolve
        current_beta = mu['beta']
        mat_ng = self.jacobian_func(vec_ng, current_beta)
        
      
        return NGSolveMatrixOperator(mat_ng, range=self.range, source=self.source)

```

```python
# ingridients for the Model

# create the concrete operator we want (that works in PyMOR but uses ng computations)
a_br = BrusselatorOperator(
    V_h=V_h, 
    a_form=a_op, 
    residual_func=spatial_residual, 
    jacobian_func=jacobian
)


# take some function as initial cond
gfu0 = GridFunction(V_h)
gfu0.components[0].Set(0.1 * cos(1.57 * x))
gfu0.components[1].Set(0.1 * sin(3.14 * x)) 
u0_vec = v_space.make_array([gfu0])
```

```python
# Test Mass Matrix 
try:
    test_apply = m_op.apply(u0_vec)
    print("Mass matrix works")
except AttributeError as e:
    print(f"Mass matrix failed: {e}")

# Test Operator 
try:
    test_nonlinear = a_br.apply(u0_vec, mu={'beta': 1.0})
    print("Nonlinear operator works")
except AttributeError as e:
    print(f"Nonlinear operator failed: {e}")
```

```python
# here first create the FOM with the random initial condition from above

from pymor.models.basic import InstationaryModel
from pymor.algorithms.timestepping import ImplicitEulerTimeStepper

# this sets the newton-tol for every time step
stepper = ImplicitEulerTimeStepper(nt=64)   #,solver_options={'rtol': 1e-6, 'atol': 1e-8, 'maxiter': 50  }) # take 64, as in the paper
time=9                             # time intervals for one period go from ~6 to ~9, here we first make it parameter-independent 

first_Brusselator_fom = InstationaryModel(    # here create first FOM (we will change initial cond and T in the following)
    T=time,
    initial_data=u0_vec,
    operator=a_br, # A   
    mass=m_op,     # M
    rhs=None,     # f
    time_stepper=stepper,
    products={
        'l2': m_op,          
        'h1_0_semi': H01_op,  
        'h1': m_op+H01_op          
    }
)
```

```python
# here make initial condition an operator (let the FOM run for a longer time and take last time step)
# did this manually for every parameter mu = [2.75,3.25,3.75,4.25] (and 3.0,3.5 to test against)

mu=2.75

t_long = 30.0

stepper_long = ImplicitEulerTimeStepper(nt=300,solver_options={'rtol': 1e-6, 'atol': 1e-8, 'maxiter': 200})

fom_long_for_initial = first_Brusselator_fom.with_(T=t_long, time_stepper=stepper_long)

u_long_for_initial = fom_long_for_initial.solve(mu)    # FOM but with a long time interval
periodic_u0 = u_long_for_initial[-1]    # last time step

Brusselator_fom = first_Brusselator_fom.with_(initial_data=periodic_u0)  # FOM but with new initial cond

Brusselator_fom_for_T = Brusselator_fom.with_(T=t_long, time_stepper=stepper_long) # FOM but with new initial cond, but for long time
```

```python
#  here we want to get the periods for a specific parameter

u_fom = Brusselator_fom_for_T.solve(mu) 
norms = u_fom.norm(m_op)
```

```python
# we want to know the exact periods for the parameters 2.75, 3.25, 3.75, 4.25 

# compute L2-norm of solution for given parameters, then take the highest 3 values and take the difference of the last 2

from scipy.signal import find_peaks

times = np.linspace(0, t_long, 300 + 1)
peak_indices, _ = find_peaks(norms)
peak_values = norms[peak_indices]
peak_times = times[peak_indices]

top_indices = np.argsort(peak_values)[-3:]       # take the highest norm values (take 4 for the first 2 parameter, 3 for the last)
top_peak_times = peak_times[top_indices]
sorted_top_times = np.sort(top_peak_times)

T_per = sorted_top_times[-1] - sorted_top_times[-2]

parameter_study[mu] = {
    'period': T_per,
    'initial_data': periodic_u0
}

print("Currently safed:")
for m, data in parameter_study.items():
    print(f"T_{m} = {data['period']:.4f}")

# now we have all the data we want collected in parameter_study
```

```python
# here we compute our snapshots with customized FOMs

import gc

sample_mus = [2.75, 3.25, 3.75, 4.25]
# instead of looping over these mu, compute snapshots for these 4 manually

target_intervals = 64 

#for mu in sample_mus:
mu=4.25
factor = 2
internal_nt = target_intervals * factor    # so that newton converges, but we just take every /factor element into our snapshots
        
T_mu = parameter_study[mu]['period']        # to get the corresponding initial data and final time T from our list 
u0_raw_np = parameter_study[mu]['initial_data'].to_numpy().flatten()
u0_mu = v_space.zeros(1)

gfu_impl = u0_mu.vectors[0].real_part.impl
gfu_impl.vec.FV().NumPy()[:] = u0_raw_np
u0_single = u0_mu[0]

print(f"{mu} (T = {T_mu})...")

fom_mu = first_Brusselator_fom.with_(   # customized FOM
    T=T_mu, 
    initial_data=u0_single, 
    time_stepper=ImplicitEulerTimeStepper(nt=internal_nt)
    )
    
U = fom_mu.solve(mu)
#U = U_fine[::factor]
std_snapshots.append(U)

#del U_fine
del U
gc.collect()

len(std_snapshots)  # should be 260
```

```python
# create curly U snapshots, customized version

import numpy as np

U_curly = Brusselator_fom.solution_space.empty() 

sample_mus = [2.75, 3.25, 3.75, 4.25]
L = len(sample_mus) - 1
M = 64*2 
N = (L + 1) * (M + 1)
d_alpha = sample_mus[1] - sample_mus[0]

all_solutions = []
all_dts = []    # since we have different final times, we have different dt aswell

for i, mu in enumerate(sample_mus):

    start = i * (M + 1)
    end = (i + 1) * (M + 1)
    
    sol_mu = std_snapshots[start:end].copy()   # we take the solutions we already computed in std_snapshots
   
    raw_data = sol_mu.to_numpy()            # this is just to make the pickle data compatible
    sol_mu = Brusselator_fom.solution_space.from_numpy(raw_data)
    
    all_solutions.append(sol_mu)
    
    T_mu = parameter_study[mu]['period']  
    all_dts.append(T_mu / M)

for l in range(L + 1):
    vector = all_solutions[l][0].copy()
    vector *= np.sqrt(N)
    U_curly.append(vector)

for j in range(1, M + 1):
    dt_0 = all_dts[0]
    diff_t = (all_solutions[0][j] - all_solutions[0][j-1]) * (1.0 / dt_0)
    diff_t *= np.sqrt(L + 1)
    U_curly.append(diff_t)

for l in range(1, L + 1):
    dt_l = all_dts[l]
    for j in range(1, M + 1):
        u_alpha_diff_j = (all_solutions[l][j] - all_solutions[l-1][j]) * (1.0 / d_alpha)
        u_alpha_diff_j_minus_1 = (all_solutions[l][j-1] - all_solutions[l-1][j-1]) * (1.0 / d_alpha)
        
        mixed_diff = (u_alpha_diff_j - u_alpha_diff_j_minus_1) * (1.0 / dt_l)
        U_curly.append(mixed_diff)

print(len(U_curly), N, M, d_alpha)
```

```python
# (1) in-sample tests for customized ROMs

mu=3.25

T_test = parameter_study[mu]['period'] 
#u0_test = parameter_study[mu]['initial_data'] 

u0_raw = parameter_study[mu]['initial_data'] 
u0_test_array = v_space.zeros(1)
u0_test_array.vectors[0].real_part.impl.vec.FV().NumPy()[:] = u0_raw.to_numpy().flatten()
u0_test = u0_test_array[0]

fom_test = first_Brusselator_fom.with_(
    T=T_test, 
    initial_data=u0_test,
    time_stepper=ImplicitEulerTimeStepper(nt=2*64)
)

reductor_std_test = ParabolicRBReductor(
    fom=fom_test,
    RB=std_pod_basis,
    product=Brusselator_fom.products['h1_0_semi'],
    check_orthonormality=False
)

reductor_U_test = ParabolicRBReductor(
    fom=fom_test,
    RB=U_pod_basis,
    product=Brusselator_fom.products['h1_0_semi'],
    check_orthonormality=False
)

```

```python
r=38

#compute std ROM solution
rom_std_r = reductor_std_test.reduce(dims={'RB': r})   # create r-dim rom
u_rom_std = rom_std_r.solve(mu) # solve rom for this mu in low dim
u_rec_std = reductor_std_test.reconstruct(u_rom_std) # back to high dim
error_rom_std = (u_fom_test - u_rec_std).norm(Brusselator_fom.products['h1_0_semi'])  # (Pr(u_h) - u_r)
error_rom_std
```

```python
r=38

#compute std ROM solution
rom_std_r = reductor_std_test.reduce(dims={'RB': r})   # create r-dim rom
u_rom_std = rom_std_r.solve(mu) # solve rom for this mu in low dim
u_rec_std = reductor_std_test.reconstruct(u_rom_std) # back to high dim

# compute U ROM solution
rom_U_r = reductor_U_test.reduce(dims={'RB': r})
u_rom_U = rom_U_r.solve(mu) # solve rom for this mu in low dim
u_rec_U = reductor_U_test.reconstruct(u_rom_U) # back to high dim

# compute both ROM errors                                         # we can use this for in-sample, but not for out-of-sample parameters
# l = sample_mus.index(mu) # find index in std_snapshots
# start = l * (M + 1)
# end = (l + 1) * (M + 1)
# u_fom_test = std_snapshots[start:end]   # take reference solution from our set of std_snapshots

error_rom_U = (u_fom_test - u_rec_U).norm(Brusselator_fom.products['h1_0_semi'])  # (Pr(u_h) - u_r)
error_rom_std = (u_fom_test - u_rec_std).norm(Brusselator_fom.products['h1_0_semi'])  # (Pr(u_h) - u_r)

# for the error anylysis: (u_h - u_r) = (u_h - Pr(u_h)) + (Pr(u_h) - u_r))
# compute Pr(u_h), the projection of the FOM solution onto pod_basis (with r modes)

# for U
coeffs_U = Brusselator_fom.products['h1_0_semi'].apply2(U_pod_basis[:r], u_fom_test) 
u_projected_U = U_pod_basis[:r].lincomb(coeffs_U)                       # proj onto U ROM
error_proj_U = (u_fom_test - u_projected_U).norm(Brusselator_fom.products['h1_0_semi']) # (u_h - Pr(u_h))

# for std
coeffs_std = Brusselator_fom.products['h1_0_semi'].apply2(std_pod_basis[:r], u_fom_test) 
u_projected_std = std_pod_basis[:r].lincomb(coeffs_std)              # proj onto std ROM
error_proj_std = (u_fom_test - u_projected_std).norm(Brusselator_fom.products['h1_0_semi']) # (u_h - Pr(u_h)) max over time

```

```python
def extract_time_series(u_array, x_val=0.5, current_mu=3.0, alpha_val=1.0):
    y_vals = []
    z_vals = []
    
    # Dummy GridFunction für die Auswertung erstellen
    gf = GridFunction(V_h)
    
    for i in range(len(u_array)):
        # Jeden Zeitschritt einzeln extrahieren und flachklopfen
        gf.vec.FV().NumPy()[:] = u_array[i].to_numpy().flatten()
        
        # GridFunction-Komponenten EINZELN am Punkt x_val auswerten
        val_u = gf.components[0](mesh(x_val))
        val_v = gf.components[1](mesh(x_val))
        
        # Physikalische Werte rekonstruieren (Offset addieren!)
        y_real = alpha_val + val_u
        z_real = (current_mu / alpha_val) + val_v
        
        y_vals.append(y_real)
        z_vals.append(z_real)
        
    return np.array(y_vals), np.array(z_vals)
# 2. Daten extrahieren
y_fom, z_fom = extract_time_series(u_fom_test, current_mu=mu)
y_std, z_std = extract_time_series(u_rec_std, current_mu=mu)
y_U, z_U = extract_time_series(u_rec_U, current_mu=mu)

# 3. Plot 1: Zeitreihen von y und z am Punkt x=0.5
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(times, y_fom, 'k-', label='FOM')
plt.plot(times, y_std, 'r--', label='Std ROM')
plt.plot(times, y_U, 'b-.', label='New ROM')
plt.title('Zeitreihe y(x=0.5, t)')
plt.xlabel('Zeit')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(times, z_fom, 'k-', label='FOM')
plt.plot(times, z_std, 'r--', label='Std ROM')
plt.plot(times, z_U, 'b-.', label='New ROM')
plt.title('Zeitreihe z(x=0.5, t)')
plt.xlabel('Zeit')
plt.ylabel('z')
plt.legend()

plt.tight_layout()
plt.show()

# 4. Plot 2: Phasenraum-Portrait (y gegen z)
plt.figure(figsize=(7, 6))

plt.plot(y_fom, z_fom, 'k-', label='FOM Limit Cycle')
plt.plot(y_std, z_std, 'r--', label='Std ROM')
plt.plot(y_U, z_U, 'b-.', label='New ROM')

# Startpunkt markieren, um die Richtung zu sehen
plt.plot(y_fom[0], z_fom[0], 'go', label='Start (t=0)')

plt.title('Phasenraum-Portrait bei x=0.5')
plt.xlabel('y')
plt.ylabel('z')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# this is for the not-customized FOM

# compute both ROMs for one test parameter and compute errors (here everything goes for fixed r)
# (unterschied zu unten: kein loop über r, kein loop über parameter, kein max über time)

r= 11

test_mu=3.0

# compute fom solution
u_fom_test1 = Brusselator_fom.solve(test_mu)

#compute std ROM solution
rom_std_r = reductor_std.reduce(dims={'RB': r})   # create r-dim rom
u_rom_std = rom_std_r.solve(test_mu) # solve rom for this mu in low dim
u_rec_std = reductor_std.reconstruct(u_rom_std) # back to high dim

# compute U ROM solution
rom_U_r = reductor_U.reduce(dims={'RB': r})
u_rom_U = rom_U_r.solve(test_mu) # solve rom for this mu in low dim
u_rec_U = reductor_U.reconstruct(u_rom_U) # back to high dim

# compute both ROM errors
error_rom_U = (u_fom_test1 - u_rec_U).norm(Brusselator_fom.products['h1_0_semi'])  # (Pr(u_h) - u_r)
error_rom_std = (u_fom_test1 - u_rec_std).norm(Brusselator_fom.products['h1_0_semi'])  # (Pr(u_h) - u_r)

# for the error anylysis: (u_h - u_r) = (u_h - Pr(u_h)) + (Pr(u_h) - u_r))
# compute Pr(u_h), the projection of the FOM solution onto pod_basis (with r modes)

# for U
coeffs_U = Brusselator_fom.products['h1_0_semi'].apply2(U_pod_basis[:r], u_fom_test1) 
u_projected_U = U_pod_basis[:r].lincomb(coeffs_U)                       # proj onto U ROM
error_proj_U = (u_fom_test1 - u_projected_U).norm(Brusselator_fom.products['h1_0_semi']) # (u_h - Pr(u_h))

# for std
coeffs_std = Brusselator_fom.products['h1_0_semi'].apply2(std_pod_basis[:r], u_fom_test1) 
u_projected_std = std_pod_basis[:r].lincomb(coeffs_std)              # proj onto std ROM
error_proj_std = (u_fom_test1 - u_projected_std).norm(Brusselator_fom.products['h1_0_semi']) # (u_h - Pr(u_h)) max over time

```

```python
# now we consider different values for r and max over t, still one parameter, std ROM

import gc # garbage

max_r = 22
r_values = range(1, max_r + 1, 5) 
errors_std = []
errors_proj_std = []
test_mu = 3.5  #[2.75,4.25], 3.0, 2.75, 3.25
u_fom_test1 = Brusselator_fom.solve(test_mu)

for r in r_values:

        print(f"start{r}")
 
        rom_r_std = reductor_std.reduce(dims={'RB': r})  # construct ROM with r modes
        u_rom_std = rom_r_std.solve(test_mu) # solve for testmu
        u_rec_std = reductor_std.reconstruct(u_rom_std) # reconstruct to high dim
    
        err = np.max((u_fom_test1 - u_rec_std).norm(Brusselator_fom.h1_0_semi_product))   # consider error max{t}\norm{(u_h - u_r^std)}
        errors_std.append(err)

        # for std
        coeffs_std = Brusselator_fom.products['h1_0_semi'].apply2(std_pod_basis[:r], u_fom_test1) 
        u_projected_std = std_pod_basis[:r].lincomb(coeffs_std)              # proj onto std ROM
    
        error_proj_std_max = np.max((u_fom_test1 - u_projected_std).norm(Brusselator_fom.products['h1_0_semi'])) # (u_h - Pr(u_h)) max over t
        
        errors_proj_std.append(error_proj_std_max)
    
        filename_err_std = f"errors_std_mu_{test_mu}.npy"
        filename_proj_std = f"proj_std_mu_{test_mu}.npy"

        np.save(filename_err_std, np.array(errors_std))
        np.save(filename_proj_std, np.array(errors_proj_std))
    
        del u_rec_std, u_rom_std
        del u_projected_std
        gc.collect()
        
plt.figure(figsize=(10, 6))
plt.semilogy(r_values, errors_std, 'b-o', label='std ROM error')
plt.semilogy(r_values, errors_proj_std, 'r--o', label='Proj. Error (Std)')
plt.xlabel('modes r')
plt.ylabel('$max_{t}$ error (H1-Seminorm)')
plt.title(fr"Convergence Std POD-ROM ($\beta$={test_mu})")
#plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()
```

```python
# now we consider different values for r and max over t, still one parameter, curly U ROM

import gc 
max_r_U = 22
r_values_U = range(1, max_r_U + 1, 5)

errors_U = []
errors_proj_U = []
test_mu = 3.5  #[2.75,4.25], 3.0, 2.75, 3.25
u_fom_test1 = Brusselator_fom.solve(test_mu)

for r in r_values_U:

        print(f"start{r}")

        rom_r_U = reductor_U.reduce(dims={'RB': r}) # construct ROM with r modes
        u_rom_U = rom_r_U.solve(test_mu) # solve for testmu
        u_rec_U = reductor_U.reconstruct(u_rom_U) # reconstruct to high dim
    
       
        err = np.max((u_fom_test1 - u_rec_U).norm(Brusselator_fom.h1_0_semi_product))   # consider error max{t}\norm{(u_h - u_r^std)}
        errors_U.append(err)

        
        # for U
        coeffs_U = Brusselator_fom.products['h1_0_semi'].apply2(U_pod_basis[:r], u_fom_test1) 
        u_projected_U = U_pod_basis[:r].lincomb(coeffs_U)              # proj onto U ROM
    
        error_proj_U_max = np.max((u_fom_test1 - u_projected_U).norm(Brusselator_fom.products['h1_0_semi'])) # (u_h - Pr(u_h)) max over t
        
        errors_proj_U.append(error_proj_U_max)

        filename_err = f"errors_U_mu_{test_mu}.npy"
        filename_proj = f"proj_U_mu_{test_mu}.npy"

        np.save(filename_err, np.array(errors_U))
        np.save(filename_proj, np.array(errors_proj_U))
    
        del u_rec_U 
        del u_rom_U
        del u_projected_U
        gc.collect()
    
    
plt.figure(figsize=(10, 6))
plt.semilogy(r_values_U, errors_U, 'b-o', label='New POD-ROM Error')
plt.semilogy(r_values_U, errors_proj_U, 'r--o', label='Proj. Error (new)')
plt.xlabel('modes r')
plt.ylabel('$max_{t}$ error (H1-Seminorm)')
plt.title(fr"Convergence New POD-ROM ($\beta$={test_mu})")
#plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()
```
