```python
import pymor                # This defines the name 'pymor'
from pymor.basic import * # This gets all the pyMOR tools
import matplotlib.pyplot as plt

print(f"pyMOR is ready! Version: {pymor.__version__}")
```

```python
from pymor.models.examples import thermal_block_example

fom_tb = thermal_block_example()
```

```python
fom_tb.parameters
```

```python
mu = [0.1, 0.2, 0.5, 1]
U = fom_tb.solve(mu) # solve in high dim
fom_tb.visualize(U);
```

```python
from pymor.parameters.functionals import ExpressionParameterFunctional
from pymor.reductors.coercive import CoerciveRBReductor

reductor = CoerciveRBReductor(
    fom_tb,
    product=fom_tb.h1_0_semi_product, # which inner product to consider
    coercivity_estimator=ExpressionParameterFunctional('min(diffusion)', # parameters
                                                       fom_tb.parameters)
)
```

```python
from pymor.algorithms.greedy import rb_greedy

parameter_space = fom_tb.parameters.space(0.1, 1)
greedy_data = rb_greedy(fom_tb, reductor, parameter_space.sample_randomly(1000),
                        rtol=1e-2)
rom_tb = greedy_data['rom'] 
```

```python
help(rb_greedy)
```

```python
rom_tb;
```

```python
Urom = rom_tb.solve(mu) # solve in low dim
Urec = reductor.reconstruct(Urom) # back to high dim
fom_tb.visualize(U - Urec) # error in high dim
```

```python
from time import perf_counter
tic = perf_counter()
fom_tb.solve(mu)
toc = perf_counter()
rom_tb.solve(mu)
tac = perf_counter()
print(f't_fom: {toc-tic}  t_rom: {tac-toc}  speedup: {(toc-tic)/(tac-toc)}')
```

```python
from pymor.algorithms.pod import pod
help(pod)
```

```python
snapshots = fom_tb.solution_space.empty()  # leeres Vh
mus = parameter_space.sample_randomly(100)  # random snapshots
for mu in mus:
    snapshots.append(fom_tb.solve(mu)) # snapshots=(set of old basis)

basis, singular_values = pod(     # POD generates POD basis and the matrix C which eigenvalues we consider
    snapshots,
    product=fom_tb.h1_0_semi_product,
    rtol=1e-4
)
```

```python
reductor2 = StationaryRBReductor(  # galerkin proj onto our space spanned by basis
    fom_tb,
    basis,
    product=fom_tb.h1_0_semi_product
)

rom2_tb = reductor2.reduce()
```

```python
mu2 = [0.2, 0.22, 0.7, 1]
Urom2 = rom2_tb.solve(mu2) # solve in low dim
Urec2 = reductor2.reconstruct(Urom2) # back to high dim
fom_tb.visualize(U - Urec2) # POD error in high dim
```

```python
fom_tb.visualize(Urec - Urec2)
```
