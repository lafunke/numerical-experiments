```python
import numpy as np
from scipy.signal import find_peaks

# parameters we want to compute (and test against)
#mus = [2.75, 3.25, 3.75, 4.25]

mu = 2.75

# let FOM run for a longer time to reach the limit cycle
t_long = 30
fom_long = create_fom(nt=700, h=50, time=t_long) 
u_long = fom_long.solve(mu)
    
# compute L2-norm of the long solution
m_op = fom_long.mass
norms = u_long.norm(m_op)

# create times array that exactly matches the trajectory length
times = np.linspace(0, t_long, len(u_long)) 
    
# find all peaks in the norms
peak_indices, _ = find_peaks(norms)
    
# compute exact period from the difference of the last 2 peaks
T_per = times[peak_indices[-1]] - times[peak_indices[-2]]
    
# take exact state at the last peak as initial cond
last_peak_index = peak_indices[-1]
periodic_u0 = u_long[last_peak_index].to_numpy() 
    
parameterstudy[mu] = {
    'period': T_per,
    'initial_data': periodic_u0
}
    
    
print("Currently safed in parameterstudy:")
for m, data in parameterstudy.items():
    print(f"mu={m} -> T_per = {data['period']}")
```

```python
# here make initial condition an operator (let the FOM run for a longer time and take last time step)
# did this manually for every parameter mu = [2.75,3.25,3.75,4.25] (and 3.0,3.5 to test against)

mu=4.25

t_long = 30
fom_long_for_initial = create_fom(nt=700, h=50, time=t_long)
u_long_for_initial = fom_long_for_initial.solve(mu)    # compute FOM solution but with a long time interval

periodic_u0 = u_long_for_initial[-1].to_numpy()   # last time step

last_peak_index = peak_indices[-1]
periodic_u0 = u_long_for_initial[last_peak_index].to_numpy()  # take last peak as initial cond

Brusselator_fom_for_T = create_fom(nt=300, h=50, time=20, initial_array = periodic_u0)  # FOM but with new initial cond
```

```python
#  here we want to get the periods for a specific parameter

u_fom = Brusselator_fom_for_T.solve(mu) 
m_op = Brusselator_fom_for_T.mass
norms = u_fom.norm(m_op)
```

```python
# we want to know the exact periods for the parameters 2.75, 3.25, 3.75, 4.25 

# compute L2-norm of solution for given parameters, then take the highest 3 values and take the difference of the last 2

from scipy.signal import find_peaks
import numpy as np
times = np.linspace(0, 20, 300 + 1)
peak_indices, _ = find_peaks(norms)
peak_values = norms[peak_indices]
peak_times = times[peak_indices]

top_indices = np.argsort(peak_values)[-2:]       
top_peak_times = peak_times[top_indices]
sorted_top_times = np.sort(top_peak_times)

T_per = sorted_top_times[-1] - sorted_top_times[-2]

parameterstudy[mu] = {
    'period': T_per,
    'initial_data': periodic_u0
}

print("Currently safed:")
for m, data in parameterstudy.items():
    print(f"T_{m} = {data['period']:.4f}")

# now we have all the data we want collected in parameter_study (initial conditions and final times)
```
