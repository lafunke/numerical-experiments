```python
from fom import create_fom

fom = create_fom(nt= 6400, h=50, time=9)

print(fom)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ngsolve import GridFunction

V_h = fom.operator.V_h
mesh = V_h.mesh

num_x_points = 200
num_t_points = len(U_trajectory)

x_werte = np.linspace(0, 1, num_x_points)
t_werte = np.linspace(0, fom.T, num_t_points)

u_matrix = np.zeros((num_x_points, num_t_points))
v_matrix = np.zeros((num_x_points, num_t_points))

gfu_plot = GridFunction(V_h)

for t_idx in range(num_t_points):
    vec_pymor = U_trajectory[t_idx].vectors[0]
    if hasattr(vec_pymor, 'real_part'):
        vec_pymor = vec_pymor.real_part
        
    gfu_impl = vec_pymor.impl
    echter_ngs_vektor = gfu_impl.vec if hasattr(gfu_impl, 'vec') else gfu_impl
    
    gfu_plot.vec.data = echter_ngs_vektor
    
    u_matrix[:, t_idx] = [gfu_plot.components[0](mesh(x)) for x in x_werte]
    v_matrix[:, t_idx] = [gfu_plot.components[1](mesh(x)) for x in x_werte]

T, X = np.meshgrid(t_werte, x_werte)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

c1 = ax1.pcolormesh(T, X, u_matrix, cmap='viridis', shading='auto', vmin=-3.0, vmax=3.0)
ax1.set_title("u (Aktivator)")
ax1.set_xlabel("Zeit t")
ax1.set_ylabel("Ort x")
fig.colorbar(c1, ax=ax1, label="u")

c2 = ax2.pcolormesh(T, X, v_matrix, cmap='magma', shading='auto', vmin=-3.0, vmax=3.0)
ax2.set_title("v (Inhibitor)")
ax2.set_xlabel("Zeit t")
ax2.set_ylabel("Ort x")
fig.colorbar(c2, ax=ax2, label="v")

plt.tight_layout()
plt.show()
```

```python
beta = fom.parameters.parse({'beta': 3.5})
U_trajectory = fom.solve(beta)

import matplotlib.pyplot as plt
import numpy as np
from ngsolve import GridFunction
import ipywidgets as widgets
from ipywidgets import interact

V_h = fom.operator.V_h
mesh = V_h.mesh
x_werte = np.linspace(0, 1, 200)
gfu_plot = GridFunction(V_h)

def plot_brusselator_time(t_step):
    vec_pymor = U_trajectory[t_step].vectors[0]
    
    if hasattr(vec_pymor, 'real_part'):
        vec_pymor = vec_pymor.real_part
        
    gfu_impl = vec_pymor.impl
    
    if hasattr(gfu_impl, 'vec'):
        echter_ngs_vektor = gfu_impl.vec
    else:
        echter_ngs_vektor = gfu_impl

    gfu_plot.vec.data = echter_ngs_vektor

    u_werte = [gfu_plot.components[0](mesh(x)) for x in x_werte]
    v_werte = [gfu_plot.components[1](mesh(x)) for x in x_werte]

    plt.figure(figsize=(10, 5))
    plt.plot(x_werte, u_werte, label='u (Aktivator)', color='blue', linewidth=2)
    plt.plot(x_werte, v_werte, label='v (Inhibitor)', color='red', linewidth=2)
    
    plt.title(f"Brusselator: beta={beta} | Zeitschritt {t_step} / {fom.time_stepper.nt}")
    plt.xlabel("Ort x")
    plt.ylabel("Konzentration")
    plt.ylim(-3.0, 3.0)
    plt.legend(loc='upper right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

interact(
    plot_brusselator_time, 
    t_step=widgets.IntSlider(
        min=0, max=fom.time_stepper.nt, step=1, value=0, description='Zeit t:'
    )
);
```

```python
# create std snapshots 

import gc
fom.disable_caching()

parameter_space = fom.parameters.space(2.75,4.25)
number_sample_mus = 4
sample_mus = parameter_space.sample_uniformly(number_sample_mus) 

std_snapshots = fom.solution_space.empty()
for mu in sample_mus:

    U_trajectory_fine = fom.solve(mu)
    
    std_snapshots.append(U_trajectory_fine[::100])
    
    del U_trajectory_fine
    gc.collect()
    
len(std_snapshots) # nt*number_sample_mus
```

```python
# POD with std 
r_std = 50

from pymor.algorithms.pod import pod
std_pod_basis, std_pod_singular_values = pod(std_snapshots,
                                     product=fom.products['h1_0_semi'],
                                     modes=r_std  
                                    )
len(std_pod_basis)
```

```python
# construct ROM using std 

from pymor.reductors.parabolic import ParabolicRBReductor

reductor_std = ParabolicRBReductor(
       fom=fom,
       RB=std_pod_basis,
       product=fom.h1_0_semi_product,
       coercivity_estimator=None,
       check_orthonormality=False,
       check_tol=None
   )

rom_std = reductor_std.reduce()
r=25
rom_std_r = reductor_std.reduce(dims={'RB': r})   # create r-dim rom
u_rom_std = rom_std_r.solve(beta) # solve rom for this mu in low dim
U_rom_rec = reductor_std.reconstruct(u_rom_std) # back to high dim
```
